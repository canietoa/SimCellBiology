```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dt=0.001
t=0
tmax=5
x=0
xarr=[0]
tarr=[0]
lamb=1
while t<tmax:
    r=np.random.rand()
    if r<lamb*dt:
        x=1
    t+=dt
    xarr.append(x)
    tarr.append(t)
```

```python
plt.plot(tarr,xarr)
```

```python
dt=0.001
tmax=5
lamb=1
ncells=1000
dataarray=[]
for n in range(ncells):
    t=0
    x=0
    while t<tmax:
        r=np.random.rand()
        if r<lamb*dt:
            x=1
        t+=dt
        dataarray.append([n,t,x])
```

```python
df=pd.DataFrame(dataarray,columns=['Cell','Time','State'])
```

```python
for cell in df.Cell.unique()[:10]:
    dfcell=df[df.Cell==cell]
    plt.plot(dfcell.Time,dfcell.State)
```

```python
meanarray=[]
for time in df.Time.unique():
    dft=df[df.Time==time]
    mn=np.mean(dft.State)
    meanarray.append(mn)
plt.plot(df.Time.unique(),meanarray)
```

```python
theory=1-np.exp(-lamb*df.Time.unique())
plt.plot(df.Time.unique(),meanarray)
plt.plot(df.Time.unique(),theory)
plt.xlabel('Time')
plt.ylabel('Mean State')
```

```python
lamb=1
ncells=1000
dataarray=[]
for n in range(ncells):
    t=0
    x=0
    r=np.random.rand()
    tr=-(1/lamb)*np.log(r)
    dt=0.01
    tmax=5
    while t<tmax:
        if t<tr:
            x=0
        else:
            x=1
        t+=dt
        dataarray.append([n,t,x])

df=pd.DataFrame(dataarray,columns=['Cell','Time','State'])
```

```python
for cell in df.Cell.unique()[:10]:
    dfcell=df[df.Cell==cell]
    plt.plot(dfcell.Time,dfcell.State)
```
